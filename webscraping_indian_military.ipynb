# Importing Required libraries:

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# checking the status code:

```python
url = "https://en.wikipedia.org/wiki/List_of_active_Indian_military_aircraft"
response = requests.get(url)
response
```

# viewng the data on page:

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

# finding all the tables:

```python
tables = soup.find_all('table', {'class': 'wikitable'})
print(f"Found {len(tables)} tables")
```

# Scrape data from the relevant table(s)

```python
dfs=[]
for table in tables:
    df = pd.read_html(str(table))[0]
    # Keep only the first 5 columns (or all if the table has fewer than 5)
    df = df.iloc[:, :5]
    dfs.append(df)

dfs
```

# Combine tables into one DataFrame

```python
final_df = pd.concat(dfs, ignore_index=True)
final_df
```

# Preview and Save to CSV

```python
final_df.to_csv('indian_military_aircraft.csv', index=False)
```
